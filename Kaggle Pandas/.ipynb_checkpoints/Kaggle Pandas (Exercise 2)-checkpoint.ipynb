# Create a DataFrame italian_wines containing reviews of wines made in Italy. Hint: reviews.country equals what?

```python
# Q8 Create a DataFrame italian_wines containing reviews of wines made in Italy. Hint: reviews.country equals what?

reviews = pd.read_csv('winemag-data-130k-v2.csv',index_col=0)

italian_wines = reviews[reviews.country == 'Italy']

# Check your answer

italian_wines
```

# Create a DataFrame top_oceania_wines containing all reviews with at least 95 points (out of 100) for wines from Australia or New Zealand.

```python
# Q9 Create a DataFrame top_oceania_wines containing all reviews with at least 95 points (out of 100) 
#for wines from Australia or New Zealand.

top_oceania_wines = reviews.loc[
    (reviews.country.isin(['Australia', 'New Zealand']))
    & (reviews.points >= 95)
]

# Check your answer

top_oceania_wines
```
